```python
print("Enter the range for prime number:") 
a=int(input("Enter the minimum number"))  # take minimum value range
b=int(input("Enter the maximum number"))  # take maximum value
for prime_number in range(a,b+1):  # write for loop to find prime number from given range
    if prime_number>1:  # cheack number is greater than 1 if yes then execute another for loop
        for i in range (2,prime_number):  #write for loop for divide each number with prime number
            if (prime_number % i==0):  #cheack reminder is zero or not
                break  # if reminder is zero then break the loop
        else:
            print(prime_number)  # write else condition for for loop if reminder is not zero then print the number only ones.
 
```

```python
a=int(input("Enter the number for Factorial: ")) # take number
factorial=1 #set factorial variable at 1
if a==0: # cheack input number is zer or not
      print("factorial of number",a,"is",1) # if zero then print output
else:
      for number in range(1,a+1): #if not zero then set range from 1 to entered number
          factorial=factorial*number # find factorial of number
      print("factorial of number",a,"is",factorial) # print factorial
      
```
